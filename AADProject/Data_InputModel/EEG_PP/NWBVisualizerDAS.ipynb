```python
import re

eeg_proc = nwb.processing["eeg_preprocessed"]

def tid_from_key(k):
    m = re.match(r"trial_(\d+)_EEG_preprocessed$", k)
    return int(m.group(1)) if m else None

keys = sorted(eeg_proc.data_interfaces.keys(), key=lambda k: tid_from_key(k) or 10**9)

for k in keys:
    ts = eeg_proc.data_interfaces[k]
    tid = tid_from_key(k)
    dur = ts.data.shape[0] / ts.rate
    print(tid, k, ts.data.shape, dur)

```
